```python
import pickle as pk
import matplotlib.pyplot as plt
```

```python
with open('random_deeper_num_10_experiment_res.pk', 'rb') as f:
    random_res_10 = pk.load(f)
with open('random_deeper_num_100_experiment_res.pk', 'rb') as f:
    random_res_100 = pk.load(f)
with open('random_deeper_num_1000_experiment_res.pk', 'rb') as f:
    random_res_1000 = pk.load(f)
```

```python
train_acc_10 = random_res_10['Epoch acc list']
train_loss_10 = random_res_10['Epoch loss list']

train_acc_100 = random_res_100['Epoch acc list']
train_loss_100 = random_res_100['Epoch loss list']

train_acc_1000 = random_res_1000['Epoch acc list']
train_loss_1000 = random_res_1000['Epoch loss list']
```

```python
plt.plot(train_acc_10, label='num_subgraphs=10')
plt.plot(train_acc_100, label='num_subgraphs=100')
plt.plot(train_acc_1000, label='num_subgraphs=1000')

plt.xlabel('epochs')
plt.ylabel('train_acc')
plt.title('Random sampling train_acc')
plt.grid()
plt.legend()
plt.savefig('random_train_acc.png')
```

```python
plt.plot(train_loss_10, label='num_subgraphs=10')
plt.plot(train_loss_100, label='num_subgraphs=100')
plt.plot(train_loss_1000, label='num_subgraphs=1000')


plt.xlabel('epochs')
plt.ylabel('train_loss')
plt.title('Random sampling train_loss')
plt.grid()
plt.legend()
plt.savefig('random_loss.png')
```
